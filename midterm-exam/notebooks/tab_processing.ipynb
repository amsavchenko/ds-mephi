```python
import pandas as pd
from transliterate import translit
```

```python
with open("Sirena-export-fixed.tab", 'r', encoding='utf-8', newline='') as f:
    buf = f.read()
```

```python
buf = buf.split("\n")
new_buf = []
for line in buf:
    new_buf.append([line[0:60], 
                    line[60:72], 
                    line[72:84],
                    line[84:96],
                    line[96:108],
                    line[108:120],
                    line[120:132],
                    line[132:138],
                    line[138:144],
                    line[144:150],
                    line[150:168],
                    line[168:180],
                    line[180:186],
                    line[186:192],
                    line[192:198],
                    line[198:204],
                    line[204:216],
                    line[216:240],
                    line[240:276],
                    line[276:336]])
buf = new_buf.copy()
```

```python
new_buf = []
for line in buf:
    new_buf.append([x.rstrip(" ") for x in line])
```

```python
data = new_buf[1:]
cols = new_buf[0]
```

```python
df = pd.DataFrame(data, columns=cols)
```

```python
new = df["PaxName"].str.split(" ", n = 2, expand = True) 
df["LastName"] = new[0]
df["FirstName"] = new[1]
df["MiddleName"] = new[2]
df['Gender'] = None
df['DepartureCountry'] = None
df['DepartureCity'] = None
df['ArrivalCountry'] = None
df['ArrivalCity'] = None
df['AirlineName'] = None
df['LoyaltyStatus'] = None
```

```python
df = df.rename(columns={ "PaxName" : "FullName", \
             "PaxBirthDate" : "BirthDate", \
             "FlightCodeSh" : "FlightNumber", \
             "e-Ticket" : "Ticket", \
             "TravelDoc" : "Document", \
             "TrvCls" : "Class", \
             "PaxAdditionalInfo" : "AdditionalInfo", \
             "" : "FF", \
             "AgentInfo" : "AgentInfo",
             "From" : "DepartureAirport", \
             "Dest" : "ArrivalAirport" })
```

```python
def condition(x):
    return str(x).lower() == "n/a" or str(x).lower() == "" or x == None


df.BirthDate = df.BirthDate.apply(lambda x: None if condition(x) else str(x).lower())
df.DepartDate = df.DepartDate.apply(lambda x: None if  condition(x) else str(x).lower())
df.DepartTime = df.DepartTime.apply(lambda x: None if  condition(x) else str(x).lower())
df.ArrivalDate = df.ArrivalDate.apply(lambda x: None if  condition(x) else str(x).lower())
df.ArrivalTime = df.ArrivalTime.apply(lambda x: None if  condition(x) else str(x).lower())
df.FlightNumber = df.FlightNumber.apply(lambda x: None if condition(x) else str(x)[:6])
df.DepartureAirport = df.DepartureAirport.apply(lambda x: None if condition(x) else str(x))
df.ArrivalAirport = df.ArrivalAirport.apply(lambda x: None if condition(x) else str(x))
df.Code = df.Code.apply(lambda x: None if condition(x) else str(x))
df.Ticket = df.Ticket.apply(lambda x: None if condition(x) else str(x))
df.Document = df.Document.apply(lambda x: None if  condition(x) else str(x))
df.Seat = df.Seat.apply(lambda x: None if condition(x) else str(x))
df.Meal = df.Meal.apply(lambda x: None if condition(x) else str(x))
df.Class = df.Class.apply(lambda x: None if condition(x) else str(x))
df.Fare = df.Fare.apply(lambda x: None if condition(x) else str(x))
df.Baggage = df.Baggage.apply(lambda x: None if condition(x) else str(x))
df.AdditionalInfo = df.AdditionalInfo.apply(lambda x: None if condition(x) else str(x))
df.FF = df.FF.apply(lambda x: None if condition(x) else str(x)[3:])
df.AgentInfo = df.AgentInfo.apply(lambda x: None if condition(x) else str(x).lower())
df.FullName = df.FullName.apply(lambda x: None if condition(x) else str(translit(str(x), 'ru', reversed=True)).lower())
df.LastName = df.LastName.apply(lambda x: None if condition(x) else str(translit(str(x), 'ru', reversed=True)).lower())
df.FirstName = df.FirstName.apply(lambda x: None if condition(x) else str(translit(str(x), 'ru', reversed=True)).lower())
df.MiddleName = df.MiddleName.apply(lambda x: None if condition(x) else str(translit(str(x), 'ru', reversed=True)).lower())
```

```python
df.to_csv("tab_processed.csv")
```

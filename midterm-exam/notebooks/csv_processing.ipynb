```python
import pandas as pd
import os
```

```python
data = pd.read_csv("BoardingData.csv", sep=';')
```

```python
data = data.rename(columns={"PassengerFirstName" : "FirstName", \
             "PassengerSecondName" : "SecondName", \
             "PassengerLastName" : "LastName", \
             "PassengerSex" : "Gender", \
             "PassengerBirthDate" : "BirthDate", \
             "PassengerDocument" : "Document", \
             "BookingCode" : "BookingCode", \
             "TicketNumber" : "TicketNumber", \
             "Baggage" : "Baggage", \
             "FlightDate" : "Date", \
             "FlightTime" : "Time", \
             "FlightNumber" : "FlightNumber", \
             "CodeShare" : "CodeShare", \
             "Destination" : "ArrivalCity" })
```

```python
data.head()
```

```python
data['FirstName'] = data['FirstName'].apply(lambda x : str(x).lower())
data['SecondName'] = data['SecondName'].apply(lambda x : str(x).lower())
data['LastName'] = data['LastName'].apply(lambda x : str(x).lower())
data['Gender'] = data['Gender'].apply(lambda x: str(x).lower())
data['BirthDate'] = data['BirthDate'].apply(lambda x : str(x).split('/')[2] + '-' + str(x).split('/')[0] + '-' + str(x).split('/')[1])
data['Document'] = data['Document'].apply(lambda x : str(x).lower())
data['BookingCode'] = data['BookingCode'].apply(lambda x : None if str(x).lower() == 'not presented' else str(x))
data['TicketNumber'] = data['TicketNumber'].apply(lambda x : None if str(x).lower() == 'not presented' else str(x))
data['Baggage'] = data['Baggage'].apply(lambda x : None if str(x).lower() == 'none' else str(x).lower())
data['Date'] = data['Date'].apply(lambda x : str(x).lower())
data['Time'] = data['Time'].apply(lambda x : str(x).lower())
data['FlightNumber'] = data['FlightNumber'].apply(lambda x : str(x))
data['CodeShare'] = data['CodeShare'].apply(lambda x : str(x))
data['ArrivalCity'] = data['ArrivalCity'].apply(lambda x : str(x).lower())
```

```python
data.to_csv("boarding_data_processed.csv", sep=',', na_rep='')
```

# Column transformation

## Names

```python
def delete_soft_sign(name):
    pos = name.find("'")
    if pos == -1:
        return name
    return name[:pos] + name[pos+1:]
```

```python
def replace_chars(name):
    pos_x = name.find('x')
    if pos_x != -1:
        name = name[:pos_x] + 'ks' + name[pos_x+1:]
    
    pos_ia = name.find('ia')
    if pos_ia != -1:
        name = name[:pos_ia] + 'ya' + name[pos_ia+2:]
    
    pos_iu = name.find('iu')
    if pos_iu != -1:
        name = name[:pos_iu] + 'yu' + name[pos_iu+2:]
    
    pos_yi = name.find('yi')
    if pos_yi != -1:
        name = name[:pos_yi] + 'yy' + name[pos_yi+2:]
        
    pos_tc = name.find('tc')
    if pos_tc != -1:
        name = name[:pos_tc] + 'ts' + name[pos_tc+2:]
        
    pos_io = name.find('io')
    if pos_io != -1:
        name = name[:pos_io] + 'yo' + name[pos_io+2:]
        
    pos_ii = name.find('ii')
    if pos_ii != -1:
        name = name[:pos_ii] + 'iy' + name[pos_ii+2:] 
        
    pos_ja = name.find('ja')
    if pos_ja != -1:
        name = name[:pos_ja] + 'ya' + name[pos_ja+2:]
        
    pos_ij = name.find('ij')
    if pos_ij != -1:
        name = name[:pos_ij] + 'iy' + name[pos_ij+2:]
        
    pos_ju = name.find('ju')
    if pos_ju != -1:
        name = name[:pos_ju] + 'yu' + name[pos_ju+2:]
        
    return name
```

```python
tab_df['FirstName'] = tab_df['FirstName'].map(delete_soft_sign)
tab_df['FirstName'] = tab_df['FirstName'].map(replace_chars)

tab_df['LastName'] = tab_df['LastName'].map(delete_soft_sign)
tab_df['LastName'] = tab_df['LastName'].map(replace_chars)
```

```python
tab_df['Name'] = tab_df['FirstName'] + ' ' + tab_df['LastName']
```

## Birthdates & documents & genders

```python
with open('../utility_dataframes/birth_dates.pkl', 'rb') as f:
    birth_dates = pickle.load(f)

tab_df['BirthDate'] = tab_df['Name'].map(lambda x: birth_dates.get(x, np.nan))
```

```python
with open('../utility_dataframes/documents.pkl', 'rb') as f:
    documents = pickle.load(f)

tab_df['Document'] = tab_df['Name'].map(lambda x: documents.get(x, np.nan))
```

```python
with open('../utility_dataframes/genders.pkl', 'rb') as f:
    genders = pickle.load(f)

tab_df['Gender'] = tab_df['Name'].map(lambda x: genders.get(x, np.nan))
```

```python
def add_gender(last_name):
    if last_name[-1] == 'a':
        return 'F'
    if last_name[-1] in ('n', 'v', 'y'):
        return 'M'
    return np.nan
```

```python
tab_df.loc[tab_df['Gender'].isna(), 'Gender'] = tab_df.loc[tab_df['Gender'].isna(), 'LastName'].map(add_gender)
```

```python
def extract_city_country(code):
    try:
        return tuple(airports_df.loc[code])
    except KeyError:
        return (np.nan, np.nan)
```

```python
json_bonus = set(json_name_bonus['BonusProgrammNumber'])
registered_users = set()

for i in tqdm(range(all_name_bonus.shape[0])):
    row = all_name_bonus.iloc[i]
    if row['BonusProgrammNumber'] in json_bonus:
        registered_users.add(row['Name'])
```

```python
all_reset_index = excel_csv_tab_xml_json.reset_index()
```

```python
all_reset_index['Registered'] = all_reset_index['Name'].map(lambda x: int(x in registered_users))
```

```python
all_reset_index['ArrivalCity'] = all_reset_index['ArrivalCity'].map(lower_with_check)
all_reset_index['ArrivalCountry'] = all_reset_index['ArrivalCountry'].map(lower_with_check)

all_reset_index['DepartureCity'] = all_reset_index['DepartureCity'].map(lower_with_check)
all_reset_index['DepartureCountry'] = all_reset_index['DepartureCountry'].map(lower_with_check)
```
